```python
fhand = open('water_use_data', 'r')

count=0
_list = []

for text in fhand:
    if not text.startswith("#"):
        count +=1
        _list.append(text)
count
_list[0]
```

```python
_list[3]
```

```python
import re

count =2
dict_w = {}
for row in _list[2:]:
    dict_w[count] = re.findall('(.*?)s*[\t]', _list[count])
    item_ = []
    for item in dict_w[count]:
        item = item.replace('-', '0')
        item_.append(item)
    
    dict_w[count] = item_
        
    
    #print(dict_w[count])
    count += 1

```

```python
# checking that each entry in dictionary is the same length in order to add to data frame
for item in dict_w.values():
    if len(item) != 282:
        print("NOT ENOUGH ENTRIES ")
```
